```python
dataset = pd.read_excel("online_retail_II.xlsx")
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.dropna(subset="Customer ID")
```

```python
dataset.isnull().sum()
```

now check for is there any cancled data present or not

```python
cancle_order = dataset[dataset['Invoice'].astype(str).str.startswith("C")]
cancle_order.shape
```

There is 9839 cancle data present remove all

```python
dataset = dataset[~dataset["Invoice"].astype(str).str.startswith("C")]
```

```python
dataset.shape
```

convert invoicedate into to_datetime format

```python
dataset["InvoiceDate"] = pd.to_datetime(dataset["InvoiceDate"])
```

remove invoice with 0 or minus quantity and price

```python
dataset = dataset[(dataset["Quantity"] > 0) & (dataset["Price"] > 0)]
```

```python
# this make colums for revenu:-

dataset["Total_amount"] = dataset["Price"] * dataset["Quantity"]
```

```python
dataset.to_csv("final_cleaned_data.csv")
```
